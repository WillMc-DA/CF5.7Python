### Data Wrangling

```python
# Dropping eval_set column from the Orders Dataset
df_ords = df_ords.drop(columns = ['eval_set'])
```

```python
df_ords.head(40)
```

```python
#Checking to see how many entries in the days_since_prior_order column has Nan
df_ords['days_since_prior_order'].value_counts(dropna = False)
```

```python
#Renaming the order_dow column into orders_day_of_week
df_ords.rename(columns = {'order_dow' : 'orders_day_of_week'}, inplace = True)
```

```python
#Confirming change
df_ords.head()
```

```python
#Changing the data type of the order_id column to a string
df_ords['order_id'] = df_ords['order_id'].astype('str')
```

```python
#Confirm the change
df_ords['order_id'].dtype
```

```python
#Changing the data type of the user_id column to a string
df_ords['user_id'] = df_ords['user_id'].astype('str')
```

```python
#Changing the name of order_hour_of_day to make it easier to read
df_ords.rename(columns = {'order_hour_of_day' : 'orders_hour'}, inplace = False)
```

### Transposing a dataframe (df_dep)

```python
#Checking departmants dataset
df_dep.head()
```

```python
#Transpose the Departments dataset from Wide Format to Long Format
df_dep_t = df_dep.T
```

```python
#Confirm change
df_dep_t
```

```python
#Fixing the headers PT.1
#Adding an index to the left
df_dep_t.reset_index()
```

```python
#Fixing the headers PT.2
#Saving the right headers as new_header
new_header = df_dep_t.iloc[0]
```

```python
#Confirming what was saved
new_header
```

```python
#Fixing the headers PT.3
#Removing the new_header/0 row
df_dep_t_new = df_dep_t[1:]
```

```python
#Confirming the change
df_dep_t_new
```

```python
#Replacing the current header with new_header
df_dep_t_new.columns = new_header
```

```python
#Confirming the change
df_dep_t_new
```
