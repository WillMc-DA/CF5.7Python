### Cleaning up superfluous columns in the datasets

```python
#_merge is being removed so the next merge can use that column name.
orders_products_combined = orders_products_combined.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1', '_merge'])
```

```python
orders_products_combined.head(70)
```

```python
df_prods_checked = df_prods_checked.drop(columns = ['Unnamed: 0',])
```

```python
df_prods_checked.head()
```

### Merging the 2 data sets

```python
#Merging, using the product_id column
df_merge_orders2 = orders_products_combined.merge(df_prods_checked, on = 'product_id', indicator = True)
```
